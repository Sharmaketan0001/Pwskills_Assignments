Q1:- Define a double linked list?



A double linked list (often referred to as a doubly linked list) is a type of linked list in which each node contains a reference to both the next node and the previous node in the sequence. This structure allows for efficient traversal in both directions: forward and backward.

Q2:- Write a function to reverse a linked list in-place ?

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

    def reverse(self):
        prev = None
        current = self.head
        while current is not None:
            next_node = current.next  # Save next node
            current.next = prev       # Reverse current node's pointer
            prev = current            # Move prev and current one step forward
            current = next_node
        self.head = prev  # Update head to the new first node

    def display(self):
        current = self.head
        while current:
            print(current.data, end=' -> ')
            current = current.next
        print('None')

# Driver Code
ll = LinkedList()
ll.append(1)
ll.append(2)
ll.append(3)
ll.append(4)

print("Original list:")
ll.display()  # Output: 1 -> 2 -> 3 -> 4 -> None

ll.reverse()

print("Reversed list:")
ll.display()  # Output: 4 -> 3 -> 2 -> 1 -> None
```

Q3:-  Detect cycle is a linked list?

==> Detecting a cycle in a linked list is a fundamental problem in computer science, and there are several methods to solve it. The most efficient method is Floyd's Cycle-Finding Algorithm, also known as the "Tortoise and Hare" algorithm. This algorithm uses two pointers to traverse the linked list at different speeds and can detect cycles without requiring extra space.

==> Floyd’s Cycle-Finding Algorithm
Concept
==> Two Pointers:

Tortoise (slow pointer): Moves one step at a time.
Hare (fast pointer): Moves two steps at a time.
==> Cycle Detection:

If there is no cycle, the hare will reach the end of the list (None).
If there is a cycle, the hare and tortoise will eventually meet within the cycle.

Steps
==>Initialize Pointers:

Both slow and fast pointers start at the head of the list.
Traverse the List:

Move slow by one step.
Move fast by two steps.
If fast or fast.next becomes None, the list does not contain a cycle.
If slow equals fast at any point, a cycle is detected.

Q4:- Merge two sorted linked list into one 

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

    def display(self):
        current = self.head
        while current:
            print(current.data, end=' -> ')
            current = current.next
        print('None')

def merge_sorted_lists(l1, l2):
    dummy = Node(0)  
    current = dummy

    # Traverse both lists
    while l1 and l2:
        if l1.data <= l2.data:
            current.next = l1
            l1 = l1.next
        else:
            current.next = l2
            l2 = l2.next
        current = current.next

   
    if l1:
        current.next = l1
    if l2:
        current.next = l2

    return dummy.next

# Driver Code
list1 = LinkedList()
list1.append(1)
list1.append(3)
list1.append(5)

# Create second sorted linked list
list2 = LinkedList()
list2.append(2)
list2.append(4)
list2.append(6)

print("List 1:")
list1.display()  # Output: 1 -> 3 -> 5 -> None

print("List 2:")
list2.display()  # Output: 2 -> 4 -> 6 -> None

# Merge the lists
merged_head = merge_sorted_lists(list1.head, list2.head)

# Display the merged list
merged_list = LinkedList()
merged_list.head = merged_head
print("Merged List:")
merged_list.display() 
```

Q5:- Write a function to remove nth node from the end in a linked list

1->2->3->4->5->6,removing 2nd node from end will return 1->2->3->4->6

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

    def display(self):
        current = self.head
        while current:
            print(current.data, end=' -> ')
            current = current.next
        print('None')

    def remove_nth_from_end(self, n):
        dummy = Node(0)  # Create a dummy node to handle edge cases
        dummy.next = self.head
        first = dummy
        second = dummy

        # Move the `first` pointer `n+1` steps ahead
        for _ in range(n + 1):
            if first is None:
                return  # If `n` is greater than the number of nodes
            first = first.next

        # Move both pointers until `first` reaches the end
        while first:
            first = first.next
            second = second.next

        # `second.next` is the node to be removed
        second.next = second.next.next

        # Update the head if needed
        self.head = dummy.next

# Driver code
ll = LinkedList()
ll.append(1)
ll.append(2)
ll.append(3)
ll.append(4)
ll.append(5)
ll.append(6)

print("Original list:")
ll.display()  # Output: 1 -> 2 -> 3 -> 4 -> 5 -> 6 -> None

ll.remove_nth_from_end(2)

print("List after removing 2nd node from the end:")
ll.display()  # Output: 1 -> 2 -> 3 -> 4 -> 6 -> None
```

Q6:- Remove duplicates from a sorted linked list

1->2->3->3->4->4->4->5  should be changed to 1->2->3->4->5


```python
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def remove_duplicates(head):
    current = head
    
    while current and current.next:
        if current.value == current.next.value:
            # Skip the next node
            current.next = current.next.next
        else:
            # Move to the next node
            current = current.next

    return head


def print_list(head):
    while head:
        print(head.value, end="->")
        head = head.next
    print("None")


def create_linked_list(values):
    if not values:
        return None
    head = ListNode(values[0])
    current = head
    for value in values[1:]:
        current.next = ListNode(value)
        current = current.next
    return head

values = [1, 2, 3, 3, 4, 4, 4, 5]
head = create_linked_list(values)
print("Original list:")
print_list(head)

head = remove_duplicates(head)
print("List after removing duplicates:")
print_list(head)
```

Q7:- Find the intersection of the two linked lists


1->2->3->4->8->6->9  5->1->6->7  , intersection 1->6

```python
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def find_intersection(head1, head2):
    # Store nodes of the first list in a hash set
    nodes = set()
    while head1:
        nodes.add(head1)
        head1 = head1.next
    
    # Find intersection by checking nodes of the second list
    intersection = []
    while head2:
        if head2 in nodes:
            intersection.append(head2.value)
        head2 = head2.next
    
    return intersection

# Helper functions
def print_list(head):
    while head:
        print(head.value, end="->")
        head = head.next
    print("None")

def create_linked_list(values):
    if not values:
        return None
    head = ListNode(values[0])
    current = head
    for value in values[1:]:
        current.next = ListNode(value)
        current = current.next
    return head

# Example usage
list1 = create_linked_list([1, 2, 3, 4, 8, 6, 9])
list2 = create_linked_list([5, 1, 6, 7])

# Manually create intersection
# Connecting list1 with list2 at node with value 6
node_intersection = list1
while node_intersection.value != 6:
    node_intersection = node_intersection.next
list2.next.next = node_intersection

intersection = find_intersection(list1, list2)
print("Intersection:", intersection)
```

Q8:-  Rotate a linked list by k positions to the right

 1->2->3->4->8->6->9 , after rotating for 2 times Cecomes , 3->4->8->6->9->1->2


```python
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def rotate_right(head, k):
    if not head or k == 0:
        return head

    # Step 1: Compute the length of the linked list
    length = 1
    old_tail = head
    while old_tail.next:
        old_tail = old_tail.next
        length += 1
    
    # Step 2: Normalize k
    k = k % length
    if k == 0:
        return head
    
    # Step 3: Find the new tail (length - k - 1) and the new head
    new_tail_position = length - k - 1
    new_tail = head
    for _ in range(new_tail_position):
        new_tail = new_tail.next
    
    new_head = new_tail.next
    new_tail.next = None
    old_tail.next = head
    
    return new_head

# Helper functions
def print_list(head):
    while head:
        print(head.value, end="->")
        head = head.next
    print("None")

def create_linked_list(values):
    if not values:
        return None
    head = ListNode(values[0])
    current = head
    for value in values[1:]:
        current.next = ListNode(value)
        current = current.next
    return head

#Driver Code
values = [1, 2, 3, 4, 8, 6, 9]
head = create_linked_list(values)
print("Original list:")
print_list(head)

k = 2
rotated_head = rotate_right(head, k)
print(f"List after rotating by {k} positions:")
print_list(rotated_head)
```

Q9:- Add Two Numbers Represented by LinkedLists:
Given two non-empty linked lists representing two non-negative integers, where the digits are stored in
reverse order, add the two numbers and return it as a linked list.

```python
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def add_two_numbers(l1, l2):
    # Initialize dummy node and current pointer
    dummy_head = ListNode()
    current = dummy_head
    carry = 0
    
    # Traverse both linked lists
    while l1 or l2 or carry:
        # Get values from the current nodes, default to 0 if the node is None
        val1 = l1.value if l1 else 0
        val2 = l2.value if l2 else 0
        
        # Calculate the sum and the carry
        total = val1 + val2 + carry
        carry = total // 10
        digit = total % 10
        
        # Create a new node with the digit
        current.next = ListNode(digit)
        current = current.next
        
        # Move to the next nodes
        if l1:
            l1 = l1.next
        if l2:
            l2 = l2.next
    
    return dummy_head.next

# Helper functions
def print_list(head):
    while head:
        print(head.value, end="->")
        head = head.next
    print("None")

def create_linked_list(values):
    if not values:
        return None
    head = ListNode(values[0])
    current = head
    for value in values[1:]:
        current.next = ListNode(value)
        current = current.next
    return head

# Driver Code
list1 = create_linked_list([2, 4, 3]) # Represents the number 342
list2 = create_linked_list([5, 6, 4]) # Represents the number 465

# Add the two numbers
result = add_two_numbers(list1, list2)
print("Sum of the two numbers represented by linked lists:")
print_list(result)
```
